```python

from ipywidgets import interact, interactive, widgets

def caesar_cipher(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            ascii_offset = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - ascii_offset + shift) % 26 + ascii_offset)
        else:
            result += char
    return result

def update_result(text, shift, operation):
    if operation == 'Encrypt':
        result_text.value = caesar_cipher(text, shift)
    else:
        result_text.value = caesar_cipher(text, -shift)

# Create widgets
text_input = widgets.Textarea(value='Hello, World!', description='Text:')
shift_slider = widgets.IntSlider(value=3, min=1, max=25, description='Shift:')
operation_dropdown = widgets.Dropdown(options=['Encrypt', 'Decrypt'], value='Encrypt', description='Operation:')
result_text = widgets.Textarea(value='', description='Result:', disabled=True)

# Set up interactivity
interactive(update_result, text=text_input, shift=shift_slider, operation=operation_dropdown)

# Display widgets
widgets.VBox([text_input, shift_slider, operation_dropdown, result_text])
```
